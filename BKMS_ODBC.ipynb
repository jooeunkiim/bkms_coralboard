### 서울시 전기차충전소 현황

```python
# PostgreSQL 연결
conn = psycopg2.connect(connection_info)
filtered = []

try:
    # 커서(Cursor) 생성
    # 커서: 명령문(SQL) 실행, 결과의 현재 위치 표시
    cursor = conn.cursor()
    # fetchall(): 쿼리의 결과를 모두 가져오기
    cursor.execute('''SELECT 소재지도로명주소, 소재지지번주소, 위도, 경도, 완속충전가능여부, 급속충전가능여부, 완속충전기대수, 급속충전기대수, 관리업체명
                   FROM eleccar_charger WHERE 설치시도명 = '서울특별시' ''')
    result = cursor.fetchall()
    for r in result:
        row = []
        if r[1] and (r[1].split()[0].endswith('구') or r[1].split()[0].endswith('시') or r[1].split()[0].endswith('울')):
            if r[1].split()[0] in ['서울특별시', '서울', '서울시']:
                row.extend(r[1].split()[1:3])
            else:
                row.extend(r[1].split()[:2])
            row.append(None)
        else:
            if r[0].split()[0] in ['서울특별시', '서울', '서울시']:
                row.extend([r[0].split()[1], None, r[0].split()[2]])
            else:
                row.extend([r[0].split()[0], None, r[0].split()[1]])
        row.extend(list(r[2:]))
        filtered.append(tuple(row))
        print(row)
        
except psycopg2.Error as e:
    # 데이터베이스 에러 처리
    print("DB error: ", e)
    
finally:
    # 데이터베이스 연결 해제
    conn.close()
```

```python
# PostgreSQL 연결
conn = psycopg2.connect(connection_info)
filtered = []

try:
    # 커서(Cursor) 생성
    # 커서: 명령문(SQL) 실행, 결과의 현재 위치 표시
    cursor = conn.cursor()
    # fetchall(): 쿼리의 결과를 모두 가져오기
    cursor.execute('''
    select max(r.시군구) 구, r.읍면동명, sum(w.총인구) 총직장인구, sum(w.남성) 남성직장인구, sum(w.여성) 여성직장인구
from (select 상권_코드_명, round(avg(총_직장_인구_수)) 총인구, round(avg(남성_직장_인구_수)) 남성, round(avg(여성_직장_인구_수)) 여성
from working_population
group by 상권_코드_명) w
join (select * from road_name
where 읍면동명 IS NOT NULL
) r on w.상권_코드_명 = r.도로명
group by r.읍면동명
order by 구

    ''')
    result = cursor.fetchall()
    for r in result:
        filtered.append(r)
        print(r)
        
except psycopg2.Error as e:
    # 데이터베이스 에러 처리
    print("DB error: ", e)
    
finally:
    # 데이터베이스 연결 해제
    conn.close()
```
